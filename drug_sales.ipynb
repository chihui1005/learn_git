```python
#  解析 EMA 数据库
import json

with open('./EMA_drug_name_indication.txt','r') as file, open('./ema_database.json','w') as fl:
    file.seek(0,0)
    for line in file:
        res = {}
        drug_name = line.strip().split("\t")[0].lower()
#        indication = [line.split("\t")[1].lower()]
        Drug_name = line.strip().split("\t")[0]
        indication = line.replace(Drug_name,"").replace('"','').replace(".,",".").replace(" ,","").replace(",,",",").replace("\t","").strip().lower()
        res["drug_name"] = drug_name
        res["indication"] = indication
        res["phase"] = ""
        res["recruitment_status"] = ""
        res["company"] = ""
        res["2014_sales"] = ""
        res["2015_sales"] = ""
        res["2016_sales"] = ""
        res["2017_sales"] = ""
        res["2018_sales"] = ""
        res["sales_unit"] = "million usd"
        res["source"] = ["ema"]
        fl.write(json.dumps(res) + "\n")
        
        
        
```

```python
import json

with open('./all_database.json','r') as file, open('./fda_kegg_sider_drugbank_ttd_database.json','w') as fl:
    file.seek(0,0)
    for line in file:
        line_dict = json.loads(line.strip())
        data_source = line_dict["data_source"]
        if (data_source == "fda") or (data_source == "kegg") or (data_source == "sider") or \
            (data_source == "drug_bank") or (data_source == "ttd"):
        
            indication_list_new = []
            drug_name = line_dict["drug_name"].strip().lower()
            indication_list = line_dict["indication"]
        
            for indication in indication_list:
                indication_lower = indication.strip().lower()
                indication_list_new.append(indication_lower)        
            
            for indication in indication_list_new:
                res = {}
                res["drug_name"] = drug_name
                res["indication"] = indication.lower().strip()
                res["phase"] = ""
                res["recruitment_status"] = ""
                res["company"] = ""
                res["2014_sales"] = ""
                res["2015_sales"] = ""
                res["2016_sales"] = ""
                res["2017_sales"] = ""
                res["2018_sales"] = ""
                res["sales_unit"] = "million usd"
                res["source"] = [data_source]
                fl.write(json.dumps(res) + "\n")
            
```

```python
#   drugananyst 数据库
import json

with open('./drugananyst.txt','r') as file, open('./drugananyst_database.json','w') as fl:
    file.seek(0,0)
    num = 0
    for line in file:
        num = num + 1
#        print(num)

        drug_name = line.strip().split("\t")[0].lower()
        indication_list = line.strip().split("\t")[1].split("/")    # 是个用“/”分隔的 list
        company = line.strip().split("\t")[2].lower()
        sales_2017 = line.split("\t")[3].lower()
        sales_2018 = line.split("\t")[4].replace("\n","").lower()
        
        for indication in indication_list:
            res = {}
            res["drug_name"] = drug_name
            res["indication"] = indication.lower().strip()
            res["phase"] = ""
            res["recruitment_status"] = ""
            res["company"] = company
            res["2014_sales"] = ""
            res["2015_sales"] = ""
            res["2016_sales"] = ""
            res["2017_sales"] = sales_2017
            res["2018_sales"] = sales_2018
            res["sales_unit"] = "million usd"
            res["source"] = ["drugananyst"]
            fl.write(json.dumps(res) + "\n")
        
        
```

```python
#   clinicaltrails 数据库
import json

with open('./clinicaltrails.json','r') as file, open('./clinicaltrails_database.json','w') as fl:
    file.seek(0,0)
    for line in file:
        line_dict = json.loads(line.strip())
        drug_name_list = line_dict["Intervention"]
        indication_list = line_dict["Condition"].strip().split("\n")
        phase = line_dict["Study Phase"]
        if phase:
            phase = phase.lower()
        else:
            phase = ""
        recruitment_status = line_dict["Recruitment Status"].lower().strip()
        
        res = {}
        res["phase"] = phase
        res["recruitment_status"] = recruitment_status
        res["company"] = ""
        res["2014_sales"] = ""
        res["2015_sales"] = ""
        res["2016_sales"] = ""
        res["2017_sales"] = ""
        res["2018_sales"] = ""
        res["sales_unit"] = "million usd"
        res["source"] = ["clinicaltrails"]
        for drug_name in drug_name_list:
            for indication in indication_list:                
                res["drug_name"] = drug_name.lower().strip()
                res["indication"] = indication.lower().strip()

                fl.write(json.dumps(res) + "\n")
            
    
```

```python
#  pharmcube  医药魔方数据库

import json
with open('./pharmcube.txt','r') as file, open('./pharmcube_database.json','w') as fl:
    
    dict_qu = {}
    file.seek(0,0)
    for line in file:
        line_list = line.strip().split("\t")
        drug_name = line_list[0].strip().lower()
        company_name = line_list[1].strip().lower()
        sales_2014 = line_list[2].strip().lower()
        if sales_2014 == "--":
            sales_2014 = ""
        sales_2015 = line_list[3].strip().lower()
        if sales_2015 == "--":
            sales_2015 = ""
        sales_2016 = line_list[4].strip().lower()
        if sales_2016 == "--":
            sales_2016 = ""
        sales_2017 = line_list[5].strip().lower()
        if sales_2017 == "--":
            sales_2017 = ""
        sales_2018 = line_list[6].strip().lower()
        if sales_2018 == "--":
            sales_2018 = ""
        
        res = {}
        res["drug_name"] = drug_name
        res["indication"] = ""
        res["phase"] = ""
        res["recruitment_status"] = ""
        res["company"] = company_name
        res["2014_sales"] = sales_2014
        res["2015_sales"] = sales_2015
        res["2016_sales"] = sales_2016
        res["2017_sales"] = sales_2017
        res["2018_sales"] = sales_2018
        res["sales_unit"] = "million " + line_list[7].strip().lower()
        res["source"] = "pharmcube"
        res["phase_front_display"] = ""
        dict_qu[json.dumps(res)] = 1
    for key in dict_qu:
        fl.write(key + "\n")
        
        
        
```

```python
#  做归一、去重处理
import json
import pickle

with open('./medical_entity_dict_v5.0.pkl','rb') as file:
    med_synonyms_name_dict, med_name_cate_dict = pickle.loads(file.read())

with open('./all_eight_database.json','r') as file, open('./all_eight_database_after_guiyi_quchong.json','w') as fl:
    file.seek(0,0)
    
    dict_qu = {}
    drug_name_in_dict = 0
    for line in file:
        line_dict = json.loads(line.strip())
        drug_name = line_dict["drug_name"]
        phase = line_dict["phase"]
        
        if drug_name in med_synonyms_name_dict:
            drug_name_in_dict += 1
            line_dict["drug_name"] = med_synonyms_name_dict[drug_name]
        if "phase 4" in phase:
            phase_front_display = "approved"
        else:
            phase_front_display = phase
        
        line_dict["phase_front_display"] = phase_front_display    
        dict_qu[json.dumps(line_dict)] = 1
    for key in dict_qu:
        fl.write(key + "\n")
    
    
```

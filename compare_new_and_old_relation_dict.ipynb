```python
import json
import os

#  统计每个关系库贡献的关系对数量
def relational_database_contribute_relation_amount(file_name):
    print("start stat relational database contribute relation amount and entity amount:")
    
    with open('./all_relation_dict_file_v2.0/quchong_relation_dict_v2.1/' + file_name,'r') as file:
#        print(file_name + ":" + "\n")
        e1_cate = file_name.split("_")[0]
        e2_cate = file_name.split("_")[1]
        e1_set = set()
        e2_set = set()
        source_dict = {}
        file.seek(0,0)
        for line in file:
            line_dict = json.loads(line.strip())
            e1 = line_dict["e1"].lower()
            e2 = line_dict["e2"].lower()
            sources = line_dict["sources"]
            dict_num = {'-1': 'ttd', '-2': 'kegg', '-3': 'dgidb', '-4': 'drugbank', '-5': 'brenda', '-6': 'uniprot', '-7': 'chembl', '-8': 'drugcentral', '-9': 'malacards', '-10': 'ctd', '-11': 'biogrid', '-12': 'ebi', '-13': 'omim', '-14': 'disgenet', '-15': 'gene_organizer'}
            e1_set.add(e1)
            e2_set.add(e2)
            for source in sources:
                source = dict_num[source]       #  将数据库的编号转化为数据库英文名称
                if source in source_dict:
                    source_dict[source] += 1
                else:
                    source_dict[source] = 1
        print(e1_cate + ":" + str(len(e1_set)) + "\n" + e2_cate + ":" + str(len(e2_set)) + "\n")
        for k,v in source_dict.items():
            print(k + ":" + str(v))
    return source_dict
#  统计相较上一版本每种关系新增的关系对的数量，其中有新增的关系种类
def newly_add_relation_amount_compare_with_old(file_name):
    print("start stat newly add relation amount compare with old version:")
    
    try:
        with open('./all_relation_dict_file_v2.0/quchong_relation_dict_v2.1/' + file_name,'r') as file, \
            open('./quchong_after_guiyi_v1.0/' + file_name,'r') as fl:   #  应该是10个文件
            e1_e2_combine_list_new = set()
            e1_e2_combine_list_old = set()
            newly_add_relation_amount = 0
            fl.seek(0,0)
            for line in fl:
                line_dict = json.loads(line.strip())
                e1 = line_dict["e1"].lower()
                e2 = line_dict["e2"].lower()
                e1_e2_combine = e1 + "####" + e2
                e1_e2_combine_list_old.add(e1_e2_combine)
                
            file.seek(0,0)
            for line in file:
                line_dict = json.loads(line.strip())
                e1 = line_dict["e1"].lower()
                e2 = line_dict["e2"].lower()
                e1_e2_combine = e1 + "####" + e2
#                e1_e2_combine_list_new.add(e1_e2_combine)
                if e1_e2_combine not in e1_e2_combine_list_old:
                    newly_add_relation_amount += 1
            print("newly add relation amount is:" + str(newly_add_relation_amount) + "\n")

    except:
        print("This is a new relation......")
        with open('./all_relation_dict_file_v2.0/quchong_relation_dict_v2.1/' + file_name,'r') as file:
            e1_e2_combine_list_new = set()
            file.seek(0,0)
            for line in file:
                line_dict = json.loads(line.strip())
                e1 = line_dict["e1"].lower()
                e2 = line_dict["e2"].lower()
                e1_e2_combine = e1 + "####" + e2
                e1_e2_combine_list_new.add(e1_e2_combine)
            print("relation amount is:" + str(len(e1_e2_combine_list_new)) + "\n")
            
            
if __name__=='__main__':

#    file_folder_path_old = r"./"
    file_folder_path_new = r"./all_relation_dict_file_v2.0/quchong_relation_dict_v2.1"
    file_path_list_new = []

    for file_path_root,_,file_name_list in os.walk(file_folder_path_new):
        for file_name in file_name_list:
            if ".dict" in file_name:
                file_path_list_new.append(file_name)
                
    all_file_source_dict = {}
    num = 0
    for file_name in file_path_list_new:
        num = num + 1
        print(str(num) + ":" + file_name)
        source_dict = relational_database_contribute_relation_amount(file_name)                   #  每个关系库贡献的关系数量
        for k,v in source_dict.items():
            if k in all_file_source_dict:
                all_file_source_dict[k] = all_file_source_dict[k] + v
            else:
                all_file_source_dict[k] = v
        newly_add_relation_amount_compare_with_old(file_name)
    
    print("********Summary*****relational database contribute relation amount*******")
    for k,v in all_file_source_dict.items():
        print(k + ":" + str(v))
        
        
```
